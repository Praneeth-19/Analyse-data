```python
import matplotlib.pyplot as plt
```

```python
states = ['TX', 'CAL', 'NY']
own_website_counts = [10, 8, 12]
free_google_site_counts = [5, 3, 2]
no_website_counts = [2, 0, 4]
```

```python
plt.figure(figsize=(9,5))
plt.bar(states, own_website_counts, label='Own Website')
plt.bar(states, free_google_site_counts, bottom=own_website_counts, label='Free Google Site')
plt.bar(states, no_website_counts, bottom=[i + j for i, j in zip(own_website_counts, free_google_site_counts)], label='Website Not Available')
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Website Availability by State of Furniture Stores')
plt.legend()
plt.savefig('chart.png')
plt.show()
```
